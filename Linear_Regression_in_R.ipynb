Create the model with the lm() funtion (lm : linear model)

```
%%R
model = lm(speed ~ dist, data=cars)
model
```

```
%%R
plot(model)
abline(model)
```

```
%%R
model$coefficients
```

## Manual calculation to find the corresponding value to the distance of 30 feet:



```
%%R
model$coefficients[1] + model$coefficients[2]*30
```

## Or with the R function:

```
%%R
predict(model, data.frame(dist=30))
```
